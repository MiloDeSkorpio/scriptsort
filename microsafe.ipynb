```python

# imporatr las librerias necesarias para le funcionamiento
import os
import pandas as pd
from openpyxl import Workbook
# Directorio donde se encuentran los archivos CSV y la cual sera la base de trabajo del Script
ruta_guardado = "Microsafe"

# Definir el nombre del archivo que contendra la union de todos
validadores = "porcentrajes.xlsx"
# Definir los nombres de los archivos que se van a leer
septiembre = [
    "ORT_VALIDACIONES_1RA_QNA_SEPTIEMBRE_2023_MICROSAFE.csv",
    "ORT_Validaciones_2da_septiembre_2023_Microsafe.csv"
]
agosto = [
    "ORT_2DA_QNA_AGOSTO_2023_Microsafe.csv",
    "ORT_1RA_AGO_2023_MICROSAFE.csv"
]
julio = [
    "ORT_2DA_QNA_JUNIO_2023_Microsafe.csv",
    "ORT_VAL_2DA_QNA_JULIO_2023_MICROSAFE.csv"
]
# Lista para almacenar los DataFrames de los archivos
arc_sep = []
arc_ago = []
arc_jul = []

# Leer los archivos y almacenarlos en la lista de arc_sept
for archivo in septiembre:
    archivo_path = os.path.join(ruta_guardado, archivo)
    df = pd.read_csv(archivo_path,low_memory=False)
    arc_sep.append(df)

df_sep = pd.concat(arc_sep,ignore_index=True)
#
# Leer los archivos y almacenarlos en la lista de arc_sept
for archivo in agosto:
    archivo_path = os.path.join(ruta_guardado, archivo)
    df = pd.read_csv(archivo_path,low_memory=False)
    arc_ago.append(df)

df_ago = pd.concat(arc_ago,ignore_index=True)
# Leer los archivos y almacenarlos en la lista de arc_sept
for archivo in julio:
    archivo_path = os.path.join(ruta_guardado, archivo)
    df = pd.read_csv(archivo_path,low_memory=False)
    arc_jul.append(df)

df_jul = pd.concat(arc_jul,ignore_index=True)
#
miit_s = df_sep[df_sep['LINEA'] == "1"]
miit_s['TIPO_TRANSACCION'].value_counts()
mts = miit_s['TIPO_TRANSACCION'].value_counts()
print(mts)
#
miit_a = df_ago[df_ago['LINEA'] == "1"]
miit_a['TIPO_TRANSACCION'].value_counts()
mta = miit_a['TIPO_TRANSACCION'].value_counts()
print(mta)
#
miit_j = df_jul[df_jul['LINEA'] == "1"]
miit_j['TIPO_TRANSACCION'].value_counts()
mtj = miit_j['TIPO_TRANSACCION'].value_counts()
print(mtj)
#Resultados Transacciones 
wb = Workbook()
# Si existe una hoja llamada Sheet, se elimina para evitar crear una hoja vacia
if 'Sheet' in wb.sheetnames:
    wb.remove(wb['Sheet'])
    
hoja = wb.create_sheet(title="Porcentajes julio")
hoja2 = wb.create_sheet(title="Porcentajes agosto")
hoja3 = wb.create_sheet(title="Porcentajes septiembre")



hoja.append(['Tipo de Transacción', 'Count'])
hoja2.append(['Tipo de Transacción', 'Count'])
hoja3.append(['Tipo de Transacción', 'Count'])
for index, row in mtj.items():
    hoja.append([index, row])
for index, row in mta.items():
    hoja2.append([index, row])
for index, row in mts.items():
    hoja3.append([index, row])


prc = os.path.join(ruta_guardado,validadores)
wb.save(prc)

#
whitel_path = os.path.join(ruta_guardado,'20230911_listaBlanca_v31.csv')
white_list = pd.read_csv(whitel_path)
#print(white_list['serial_hex'])
#print(df_sep['SAM_SERIAL_HEX'])

# Crear una máscara booleana que indica si SAM_SERIAL_HEX no está en white_list['serial_hex']
mascara = df_sep['SAM_SERIAL_HEX'].isin(white_list['serial_hex'])

# Filtrar las filas de df_sep que cumplen con la condición
filas_en_df_sep_no_en_white_list = df_sep[mascara]

# Imprimir las filas que cumplen con la condición
print(filas_en_df_sep_no_en_white_list.shape)

print(df_sep.shape)
```
